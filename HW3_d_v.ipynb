```python
import pandas as pd
import numpy as np
import statsmodels.api as sm

test = pd.read_csv('test_3d_iv.csv',index_col=0)
print(test)
```

```python
X_test = test.drop(columns=['Class'])
y_test = test['Class']

train = pd.read_csv('train_3d_iv.csv',index_col=0)
X_train = train.drop(columns=['Class'])
y_train = train['Class']

print(X_train)
print(X_test)
print(y_train)
print(y_test)
```

```python
X_train_stats = sm.add_constant(X_train)
X_test_stats = sm.add_constant(X_test)

log_reg = sm.Logit(y_train,X_train_stats).fit(maxiter=15) #training on train

y_test_predicted_probab = log_reg.predict(X_test_stats) #testing on test
y_test_predicted = (y_test_predicted_probab >= 0.5).astype(int) #converting to class labels from probability
print(y_test)
print(y_test_predicted)
```

```python
mis = 0
for i in np.arange(0,y_test.shape[0]):
    if y_test_predicted[i] != y_test.iloc[i]:
        mis = mis + 1

error = (mis/(y_test.shape[0]))*100

print('The actual test error obtained is : \n',error,'%')
```
